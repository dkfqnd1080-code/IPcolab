```python

img = cv2.imread('/content/gdrive/My Drive/IP/Images/Images/lena256.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#plt.imshow(gimg1, cmap = 'gray')
cv2_imshow(gimg1)
```

```python
edgeImage = cv2.addWeighted(gimg1, 1.5, gimg1, -1, 0)
cv2_imshow(edgeImage)
```

```python
unsharpImage = cv2.addWeighted(gimg1, 2, gimg1, -1, 0)
cv2_imshow(unsharpImage)
```
