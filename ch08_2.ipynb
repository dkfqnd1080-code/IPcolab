```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
from matplotlib import pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/Images/salt.bmp')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#plt.imshow(gimg1, cmap = 'gray')
cv2_imshow(gimg1)
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(gimg1,kernel,iterations = 1)
cv2_imshow(erosion)
```

```python
opening = cv2.dilate(erosion,kernel,iterations = 1)
cv2_imshow(opening)
```

```python
dilate2 = cv2.dilate(opening,kernel,iterations = 1)
cv2_imshow(dilate2)
```

```python
erode2 = cv2.erode(dilate2,kernel,iterations = 1)
cv2_imshow(erode2)
```
