```python
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/gdrive/My Drive/IP/Images/Images/milkdrop.bmp')
gimg1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
gBin = np.zeros((gimg1.shape[0],gimg1.shape[1]), dtype=np.uint8)

threshold = 150

for i in range(gimg1.shape[0]):
  for j in range(gimg1.shape[1]):
    if gimg1[i,j] > threshold:
      gBin[i,j] = 255
cv2_imshow(gBin)
```

```python
import numpy as np
import cv2
kernel = np.ones((3,3),np.uint8)
dIlateImg = cv2.dilate(gBin,kernel,iterations = 3)
cv2_imshow(dIlateImg)
```
